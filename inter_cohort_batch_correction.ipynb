```python
import scanpy as sc
import anndata
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from scib.metrics import silhouette_batch as sb
import scib
from scib.metrics import pcr_comparison as pcr
from scib.metrics import graph_connectivity as gc
from scib.integration import combat
from scib.preprocessing import hvg_intersect
import matplotlib.pyplot as plt


# File paths
file_paths = [
    "/Users/emirhanyagmur/Desktop/brain/TMT batch/MSBB_proteomics_TMT_v3_batchready.csv",
    "/Users/emirhanyagmur/Desktop/brain/TMT batch/Rosmapr1_proteomics_TMT_v3_batchready.csv",
    "/Users/emirhanyagmur/Desktop/brain/TMT batch/Rosmapr2_proteomics_TMT_v3_batchready.csv",
    "/Users/emirhanyagmur/Desktop/brain/TMT batch/RosmapBA637_proteomics_TMT_v3_batchready.csv",
    "/Users/emirhanyagmur/Desktop/brain/TMT batch/EmoryBA9_proteomics_TMT_v3_batchready.csv",
    "/Users/emirhanyagmur/Desktop/brain/TMT batch/EmoryBA24_proteomics_TMT_v3_batchready.csv",
]

# Load datasets into a list
datasets = [pd.read_csv(file) for file in file_paths]

# Check for empty DataFrames and print the first few rows
for i, dataset in enumerate(datasets):
    if dataset.empty:
        raise ValueError(f"Dataset {i+1} is empty.")
    if dataset.isnull().values.any():
        raise ValueError(f"Dataset {i+1} contains missing values.")
    print(f"Dataset {i+1} loaded successfully:")
    print(dataset.head())
    
    # Print the missing values for each dataset
    for i, dataset in enumerate(datasets):
        missing_values = dataset.isnull().sum()
        print(f"Missing values in dataset {i+1}:")
        print(missing_values[missing_values > 0])
    

    
for i, dataset in enumerate(datasets):
    # Convert column names to lowercase
    if 'ID' in dataset.columns:
        dataset.set_index('ID', inplace=True)
    else:
        raise ValueError(f"'ID' column not found in dataset {i+1}.")
    
group_mapping = {'Control': 0, 'AD': 1}
for dataset in datasets:
    dataset['Group'] = dataset['Group'].map(group_mapping)

# Transpose and store them as AnnData objects
adatas = [sc.AnnData(X=ds) for ds in datasets]

# Assign batch labels
for i, adata in enumerate(adatas):
    adata.obs['batch'] = str(i + 1)


 # Assign the values of the group column of datasets as obsm
for adata, dataset in zip(adatas, datasets):
    adata.obs['Group'] = dataset['Group'].values
    
    # Drop the 'group' column from datasets after assigning it to obs
for dataset in datasets:
    dataset.drop(columns=['Group'], inplace=True)


# Concatenate into a single AnnData object
combined = adatas[0].concatenate(adatas[1:], batch_key='batch', join='inner')

print(combined.X.shape)
# Convert batch column to string
combined.obs['batch'] = combined.obs['batch'].astype('category')

# Check the datatype of the combined object and its attributes
print("Type of combined:", type(combined))
print("Type of combined.X:", type(combined.X))
print("Shape of combined.X:", combined.X.shape)
print("Type of combined.obs:", type(combined.obs))
print("Combined.obs head:")
print(combined.obs.head())

# Ensure combined.X is a numpy array
if not isinstance(combined.X, np.ndarray):
    raise TypeError("combined.X is not a numpy array")



# Ensure combined.obs contains batch information
if 'batch' not in combined.obs.columns:
    raise ValueError("combined.obs does not contain 'batch' column")

if np.isnan(combined.X).any():
    nan_indices = np.argwhere(np.isnan(combined.X))
    print(f"NaN values found at indices: {nan_indices}")
    raise ValueError("combined.X contains NaN values")

# Perform batch correction using Combat
combined_corrected = scib.integration.combat(combined, 'batch')


combined_corrected.obs['Group'] = combined_corrected.obs['Group'].map({0: 'Control', 1: 'AD'})
combined_corrected.write("adata_batch_corrected_trial2/TMT_brain_combined_corrected.h5ad")
combined_corrected.obs['Group'] = combined_corrected.obs['Group'].map({'Control': 0, 'AD': 1})

if not isinstance(combined_corrected.X, np.ndarray):
    raise TypeError("combined.X is not a numpy array")

np.isnan(combined_corrected.X).any(axis=0)

# Check for NaN values in combined.X
if np.isnan(combined_corrected.X).any():
    nan_indices = np.argwhere(np.isnan(combined_corrected.X))
    print(f"NaN values found at indices: {nan_indices}")
    raise ValueError("combined.X contains NaN values")

# Evaluate the correction using metrics
sc.tl.pca(combined_corrected,n_comps=50)
sc.pp.neighbors(combined_corrected,n_neighbors=30)
sc.tl.umap(combined_corrected)


sc.tl.pca(combined,n_comps=50)
sc.pp.neighbors(combined,n_neighbors=30)
sc.tl.umap(combined)   


# Graph Connectivity
graph_conn = scib.metrics.graph_connectivity(combined_corrected, label_key="batch")

# Batch ASW
batch_asw = scib.metrics.silhouette_batch(combined_corrected, batch_key="batch", label_key="Group", embed="X_pca") 

# PCR score
pcr_score = scib.metrics.pcr_comparison(combined, combined_corrected, covariate="batch")

# Dimensionality reductio
sc.tl.umap(combined_corrected)

sc.tl.umap(combined)


combined_corrected.obs['Group'] = combined_corrected.obs['Group'].map({0: 'Control', 1: 'AD'})
combined.obs['Group'] = combined.obs['Group'].map({0: 'Control', 1: 'AD'})
# Visualization for combined_corrected
sc.pl.umap(combined_corrected, color='batch', size=200,alpha=0.6 ,save="LFQ_brain_umap_corrected_batch.png")   
   

sc.pl.umap(combined, color='batch',size=200,alpha=0.6,save="LFQ_brain_umap_combined_batch.png")

# Visualization for combined_corrected by group
combined_corrected.obs.rename(columns={'Group': 'group_obs'}, inplace=True)
sc.pl.umap(combined_corrected, color='group_obs', size=200,alpha=0.6 ,save="LFQ_brain_umap_corrected_group.png")

# Visualization for combined by group
combined.obs.rename(columns={'Group': 'group_obs'}, inplace=True)
sc.pl.umap(combined, color='group_obs', size=200,alpha=0.6 ,save="LFQ_brain_umap_combined_group.png")


# Print scaled metrics
print(f"Graph Connectivity : {graph_conn}")
print(f"Batch ASW : {batch_asw}")
print(f"PCR score : {pcr_score}")
```
