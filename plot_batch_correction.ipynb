```python
import scanpy as sc
import scib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming combined is already defined and contains the data

file_paths = [
    "/Users/emirhanyagmur/Desktop/CSF/TMT batch/syn44132374_TMT_proteomics_5_batchready.csv",
    "/Users/emirhanyagmur/Desktop/CSF/TMT batch/syn20933797_cohort1_TMT_proteomics_2_batchready.csv",
    "/Users/emirhanyagmur/Desktop/CSF/TMT batch/syn20933797_cohort2_TMT_proteomics_3_batchready.csv",
    "/Users/emirhanyagmur/Desktop/CSF/TMT batch/syn30549757_neat_TMT_proteomics_3_batchready.csv",
    "/Users/emirhanyagmur/Desktop/CSF/TMT batch/syn30549757_depleted_TMT_proteomics_3_batchready.csv",
]

# Load datasets into a list
datasets = [pd.read_csv(file) for file in file_paths]

# Check for empty DataFrames and print the first few rows
for i, dataset in enumerate(datasets):
    if dataset.empty:
        raise ValueError(f"Dataset {i+1} is empty.")
    if dataset.isnull().values.any():
        raise ValueError(f"Dataset {i+1} contains missing values.")
    print(f"Dataset {i+1} loaded successfully:")
    print(dataset.head())
    
for i, dataset in enumerate(datasets):
    # Convert column names to lowercase
    if 'ID' in dataset.columns:
        dataset.set_index('ID', inplace=True)
    else:
        raise ValueError(f"'ID' column not found in dataset {i+1}.")
    
group_mapping = {'Control': 0, 'AD': 1}
for dataset in datasets:
    dataset['Group'] = dataset['Group'].map(group_mapping)

# Transpose and store them as AnnData objects
adatas = [sc.AnnData(X=ds) for ds in datasets]

# Assign batch labels
for i, adata in enumerate(adatas):
    adata.obs['batch'] = str(i + 1)


 # Assign the values of the group column of datasets as obsm
for adata, dataset in zip(adatas, datasets):
    adata.obs['Group'] = dataset['Group'].values
    
    # Drop the 'group' column from datasets after assigning it to obs
for dataset in datasets:
    dataset.drop(columns=['Group'], inplace=True)


# Concatenate into a single AnnData object
combined = adatas[0].concatenate(adatas[1:], batch_key='batch', join='inner')

print(combined.X)
# Convert batch column to string
combined.obs['batch'] = combined.obs['batch'].astype('category')

# Check the datatype of the combined object and its attributes
print("Type of combined:", type(combined))
print("Type of combined.X:", type(combined.X))
print("Shape of combined.X:", combined.X.shape)
print("Type of combined.obs:", type(combined.obs))
print("Combined.obs head:")
print(combined.obs.head())

# Ensure combined.X is a numpy array
if not isinstance(combined.X, np.ndarray):
    raise TypeError("combined.X is not a numpy array")



# Ensure combined.obs contains batch information
if 'batch' not in combined.obs.columns:
    raise ValueError("combined.obs does not contain 'batch' column")

if np.isnan(combined.X).any():
    nan_indices = np.argwhere(np.isnan(combined.X))
    print(f"NaN values found at indices: {nan_indices}")
    raise ValueError("combined.X contains NaN values")

# Function to perform batch correction and calculate metrics
def evaluate_batch_correction(combined, method, **kwargs):
    if method == 'combat':
        corrected = scib.integration.combat(combined, 'batch', **kwargs)
    elif method == 'scanorama':
        corrected = scib.integration.scanorama(combined, 'batch', **kwargs)
    elif method == 'bbknn':
        corrected = scib.integration.bbknn(combined, 'batch', **kwargs)
    else:
        raise ValueError(f"Unknown method: {method}")

    # Perform dimensionality reduction
    sc.tl.pca(corrected,n_comps=50)
    sc.pp.neighbors(corrected,n_neighbors=30, use_rep='X_pca')
    sc.tl.umap(corrected)
   
    
    sc.tl.pca(combined,n_comps=50)
    sc.pp.neighbors(combined,n_neighbors=30, use_rep='X_pca')   
    sc.tl.umap(combined)
   
    

    # Ensure there are no NaN values in the embeddings
    corrected.obsm['X_pca'] = np.nan_to_num(corrected.obsm['X_pca'])
    corrected.obsm['X_umap'] = np.nan_to_num(corrected.obsm['X_umap'])

    # Calculate metrics
    graph_conn = scib.metrics.graph_connectivity(corrected, label_key="batch")
    batch_asw_pca = scib.metrics.silhouette_batch(corrected, batch_key="batch", label_key="Group", embed="X_pca")
    batch_asw_umap = scib.metrics.silhouette_batch(corrected, batch_key="batch", label_key="Group", embed="X_umap")
    pcr_score = scib.metrics.pcr_comparison(combined, corrected, covariate="batch")

    return {
        'graph_conn': graph_conn,
        'batch_asw_pca': batch_asw_pca,
        'batch_asw_umap': batch_asw_umap,
        'pcr_score': pcr_score
    }

# Evaluate different batch correction methods
methods = ['combat', 'scanorama', 'bbknn',]
results = {}

for method in methods:
    results[method] = evaluate_batch_correction(combined, method)

# Store the results in a DataFrame
metrics_df = pd.DataFrame(results).T

# Calculate the mean of the metrics
metrics_mean = metrics_df.mean(axis=1)

# Store the mean of metrics in a variable
batch_correlation_average_score = metrics_mean

# Plot the metrics for different methods
batch_correlation_average_score.plot(kind='bar')
plt.title('TMT dataset')
plt.ylabel('Batch Correlation Average Score')
plt.xlabel('Methods')
plt.savefig('figures/LFQ_CSF_batch_correlation_average_score.png')
plt.show()

# Plot the metrics for different methods
metrics_df.plot(kind='bar')
plt.title('TMT dataset')
plt.ylabel('Metrics')
plt.xlabel('Methods')
plt.show()
```
